```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('D:/Data/indian_food.csv')
```

```python
#Checking the shape of the dataset
data.shape
print("There are {0} rows and {1} columns in the dataset".format(data.shape[0],data.shape[1]))
```

```python
data['diet'].value_counts()
```

```python
data['flavor_profile'].value_counts()
```

```python
data['course'].value_counts()
#Replacing all the -1 values in the categorical and numerical columns with nan 
data= data.replace('-1',np.nan)
data = data.replace(-1,np.nan)
```

```python
data.select_dtypes('object').head()
```

```python
#Countplot of Veg/Non-Veg
plt.subplot(1,2,1)
sns.countplot(data['flavor_profile'])
plt.title("flavour")

#Piechart for Veg/NonVeg
plt.subplot(1,2,2)
x = data['flavor_profile'].value_counts()
explode = (0, 0.0, 0, 0)
labels = ['Sweet','Spicy','Bitter','Sour']
plt.pie(x,explode=explode,labels=labels, autopct = '%1.2f%%')
plt.title("Pie Chart showing the distribution of flavor")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15,6)
plt.style.use('fivethirtyeight')

#Countplot of dish courses
sns.countplot(data['course'])
plt.title("Count Plot of course")
```

```python
plt.rcParams['figure.figsize'] = (18,6)
plt.style.use('fivethirtyeight')

#Countplot of states
sns.countplot(data['state'])
plt.title("Count Plot of state")
plt.xticks(rotation=90)
plt.show()
```

```python
a = data['state'].value_counts().reset_index()
a.sort_values(by='state',ascending=False).head(6).style.background_gradient(cmap='rainbow')
```

```python
a = data['state'].value_counts().reset_index()
a.sort_values(by='state',ascending=True).head(6).style.background_gradient(cmap='copper')
```

```python
plt.rcParams['figure.figsize'] = (15,6)
plt.style.use('fivethirtyeight')

#Countplot of region
sns.countplot(data['region'])
plt.title("Count Plot of region")
```
